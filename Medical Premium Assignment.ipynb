### We can do further analysis using machine learning models:

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
x = df.drop('PremiumPrice',axis=1)
y = df.PremiumPrice
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=43)
```

```python
Lr=RandomForestRegressor(n_estimators= 100,max_depth=10)
Lr_model=Lr.fit(x_train,y_train)
```

```python
print('Accuracy of the model on training Dataset is: ',Lr_model.score(x_train,y_train))
print ('Accuracy of the model on Test Dataset is: ',Lr_model.score(x_test,y_test))
```

```python
feature_imp = Lr_model.feature_importances_
sns.barplot(x=feature_imp, y=x.columns)
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.title("Important Features")
plt.show();
```
